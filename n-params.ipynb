```python
import paddle

def m2mb(x, million=True):
    return x * (1000000 if million else 1) * 4 / 1024 / 1024

def mb2m(x):
    return x * 1024 * 1024 / 1000000 / 4

def gflops(model, size):
    return paddle.flops(model, size) / 1e+9

def nparams(model, size):
    return paddle.summary(model, size)['total_params']
```
